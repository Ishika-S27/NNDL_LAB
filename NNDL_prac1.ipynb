```python
import numpy as np
```

```python
array_one = np.array([1,2,4])
array_two = np.array([1,3,4])
array_three = array_one + array_two
print(array_three)
```

```python
array_one = np.array([[1,2,3],[1,3,5]])
array_two = np.array([[1,3,4],[2,5,3]])
print("matrix multiplication")
array_three = np.matmul(array_one, array_two.T)
print(array_three)
```

```python
array_one = np.array([[1,2,3],[1,3,5]])
array_two = np.array([[2,3],[3,4],[7,8]])
print("matrix multiplication")
array_three = np.matmul(array_one, array_two)
print(array_three)
```
